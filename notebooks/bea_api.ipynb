```python
import requests
import pandas as pd
import os

API_KEY = os.environ.get('BEA_API_KEY')
```

```python
url = 'https://apps.bea.gov/api/data'

def fetch_bea_data(params):
    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        return data
    else:
        print(f"Failed to retrieve data: {response.status_code}")
        return None
```

```python

params = {
    'UserID': api_key,
    'method': 'GetData',
    'datasetname': 'NIPA',  # National Income and Product Accounts
    'TableName': 'T10101',  # Table for Gross Domestic Product
    'Frequency': 'Q',       # Quarterly data
    'Year': 'ALL',          # All years
    'ResultFormat': 'json'  # Response format
}

gdp_df = fetch_bea_data(params)
```
